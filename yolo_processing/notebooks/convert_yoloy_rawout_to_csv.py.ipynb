```python
import os 
import sys
os.chdir(r'F:\Research-WeiZhou\darknet\build\darknet\x64\results')
file_name = 'LK96_dTTexp1_full_behavCam.txt'
file = open(file_name)
raw_string = file.read()
```

```python
frame_strings = raw_string.split('\n\n')
keep = ['left:', 'right:', 'top', 'midpoint', 'frame_num:']
lines = [ line for frame_string in frame_strings for line in frame_string.split('\n') if line.split(' ')[0] in keep ]
```

```python
raw_nums = []
temp = []
temp.append(int(lines[0].split(':')[1]))
for line in lines[1:]:
    line_split = line[:-1].split(':')
    if line_split[0] != 'frame_num':
        temp.append(int(line_split[1]))
    else:
        raw_nums.append(temp)
        temp =[(int(line_split[1]))]
raw_nums.append(temp)       
raw_nums[-5:]
```

```python
count = 0
count2 = 0
for line in raw_nums:
    if len(line) == 6:
        count = count + 1
    elif len(line) > 6:
        if line[2] > line[7]:
            count2 = count2 + 1
            print(line)
    else:
        pass
print(count2)
count
```

```python
save = []
count_same_box = 0
for idx, line in enumerate(raw_nums):
    if len(line) == 11: 
        if (abs(line[2] - line[7]) < 30 ):
            print(line)
    if len(line) > 11:
        #print(line)
        save.append((idx, raw_nums[idx]))
```

```python
for idx, line in enumerate(raw_nums):
    if len(line) > 16:
        x1 = line[1]
        x2 = line[6]
        x3 = line[11]
        x4 = line[16]
        smallest = min(x1, x2, x3, x4)
        if x1 == small:
            line = [line[0]] + line[6:]
        elif x2 == small:
            line = line[0:6] + line[11:]
        elif x3 == small:
            line = line[0:11] + line[16:]
        else:
            line = line[0:16]
    
    if len(line) > 11:
        x1 = line[1]
        x2 = line[6]
        x3 = line[11]
        small = min (x1, x2, x3)
        if x1 == small:
            if (abs(line[12] - line[7]) < 30 ):
                line = line[0:6] + line[11:]
            else:
                line = [line[0]] + line[6:]
        elif x2 == small:
            if (abs(line[2] - line[12]) < 30 ):
                line = [line[0]] + line[6:]
            else:
                line = line[0:6] + line[11:]
        else:
            if (abs(line[2] - line[7]) < 30 ):
                line = line[0:6] + line[11:]
            else:
                line = line[0:11]
    if len(line) > 6:
        line = [line[0]] + line[2:6] + line[7:]
        if len(line) == 9 and line[1] > line[5]:
            line = [line[0]] + line[5:] + line[1:5]            
        #fix xcord to be from right to left
        line[1] = 2*line[3] - line[1]
        line[5] = 2*line[7] - line[5]
    elif len(line) > 1:
        line = [line[0]] + line[2:]
        #fix xcord to be from right to left
        line[1] = 2*line[3] - line[1]

    else:
        pass
    raw_nums[idx] = line
```
